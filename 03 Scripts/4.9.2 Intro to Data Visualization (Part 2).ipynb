# 03. Task Visualizations

## 03.1 Create a histogram of the “order_hour_of_day” column. 

```python
hist_hours = df['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram shows the frequency of each hour of the day for orders. We can see that around 10 AM, orders peak, and fall off in the evening. The y-axis appears to show the frequency of the data in millions (almost 3 million orders were made during the peak hour).

## 03.2 Create a bar chart from the “loyalty_flag” column.

```python
bar_loyalty_orders = df['loyalty_flag'].value_counts().plot.bar()
```

Even though "loyal" customers place the most orders by definition, the most orders still come from our customers with the "medium" loyalty flag.

## 03.3 Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day.

```python
# Creating new df with 30% sample and relevant columns

df_3 = small[['order_hour_of_day','prices']]
```

```python
line_prices_hours = sns.lineplot(data = df_3, x = 'order_hour_of_day',y = 'prices')
```

The graph shows that the average price of items bought spikes in the early morning hours but plunged around 10 AM.

## 03.4 Determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents

```python
# Creating new df with 30% sample and relevant columns

df_4 = small[['age', 'dependents']]
```

```python
line_age_dependents = sns.lineplot(data = df_4, x = 'age',y = 'dependents')
```

The line chart shows the relationship between age and number of dependents. There is no clear positive or negative relationship in this data. Young people seem to have on average as many dependents as older people.

## 03.5 You’ll also need to explore whether there’s a connection between age and spending power (income).

```python
df_5 = small[['age', 'income']]
```

```python
scatter_age_income = sns.scatterplot(data = df_5, x = 'age', y = 'income')
```

The scatterplot here shows that there are a lot of wage earners below the 200,000/year level. There are also a sizable chunk above the age of 40 in the 200,000-300,000 level. Higher incomes are more rare. For under-40s, they cap off around 400,000, and for over-40s they cap off around 600,000.
